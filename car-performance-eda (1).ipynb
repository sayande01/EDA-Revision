```python
df.dropna(inplace = True)
```

```python
df_cat = ['class','drive','fuel_type']
df_cont = df.select_dtypes(exclude = 'object')
```

```python
for col in df_cat:
    plt.figure(figsize=(12,6))
    sns.countplot(x = col, data = df,palette = 'Set3')
    plt.xticks(rotation = 90)
    plt.show()
```

Insights 1 : Small Sport utility cater to the most popular

```python
df['fuel_type'].value_counts().plot(kind = 'pie', autopct = '%.1f')
plt.ylabel(None)
plt.show()
```

Insights 2 : 99.6% of the vehicles are of Gas Vehicle

```python
df['year'] = df['year'].astype('object')
```

```python
for col in df_cont:
    sns.histplot(x = col, data = df, bins = 30,kde = True,color = 'skyblue')
    plt.show()
```

Insights 3 : 

city_mpg = 18

highway_mpg = 24

As in general, the Highway milage > city milage

```python
for col in df_cont:
    sns.boxplot(x = 'fuel_type', y = col, data = df, palette = 'magma' )
    plt.show()
```

Insgihts 4:

In all the cases the Diesel gives higher mean than Gas vehicles

```python
sns.heatmap(df.corr(numeric_only=True), annot = True, cmap = 'viridis', fmt = '.1f')
```

```python
sns.regplot(x = 'city_mpg', y = 'highway_mpg', data =df,color='purple')
plt.show()
```

```python
sns.regplot(x = 'cylinders', y = 'combination_mpg', data =df,color='pink')
plt.show()
```
