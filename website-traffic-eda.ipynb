# Visualization

```python
sns.histplot(df['Page Views'], bins = 15, kde = True)
plt.title('Distribution of page Views')
```

```python
df['Traffic Source'].value_counts().plot(kind = 'pie', autopct = '%.1f')
plt.ylabel(None)
plt.title('Traffic Source')
plt.show()
```

```python
df_new = df[(df['Conversion Rate'] == 1) | (df['Conversion Rate'] == 0)]
```

```python
df_new.head()
```

```python
df_new['Conversion Rate'].value_counts()
```

```python
# creating a scatter plot
plt.figure(figsize=(8,4))
sns.scatterplot(x='Time on Page', y='Conversion Rate', data=df)
plt.title('Time On Page vs Conversion Rate')
plt.xlabel('Time On Page')
plt.ylabel('Conversion Rate (in percentage)')
plt.show()
```

```python
# creating a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Session Duration', y='Bounce Rate', data=df)
plt.title('Session Duration vs Bounce Rate')
plt.xlabel('Session Duration (in minutes)')
plt.ylabel('Bounce Rate (in percentage)')
plt.show()
```
