# Visulaization

```python
df.columns
```

```python
df_cat = df.select_dtypes(include='object')
df_num = df[['Smoking_Prevalence',
       'Drug_Experimentation', 'Peer_Influence','Family_Background', 'Mental_Health',
        'Parental_Supervision','Community_Support', 'Media_Influence']]
```

```python
df_cat.head()
```

```python
df_num.head()
```

```python
for col in df_cat.columns:
  plt.figure(figsize=(10,5))
  sns.countplot(x = df_cat[col], palette = 'dark')
  plt.show()
```

```python
for col in df_num.columns:
  plt.figure(figsize=(10,5))
  sns.boxplot(x = 'Gender',data = df,y = df_num[col], palette = 'magma')
  plt.show()
```

```python
df_cont = df[['Smoking_Prevalence','Drug_Experimentation']]
```

```python
for col in df_cont.columns:
  plt.figure(figsize=(10,5))
  sns.histplot(x = col, kde = True, data =df, bins = 50,  color = 'green')
  plt.show()
```

```python
for col in df_cat.columns:
  plt.figure(figsize=(10,5))
  plt.pie(df_cat[col].value_counts(), labels = df_cat[col].value_counts().index, autopct = '%1.1f%%')
  plt.show()
```

```python
df['Year'] = df['Year'].astype('object')
```

```python
for col in df_num.columns:
  plt.figure(figsize=(10,5))
  sns.lineplot(x = 'Year', y = col, data = df, color = 'red')
  plt.show()
```
